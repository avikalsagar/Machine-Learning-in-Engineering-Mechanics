```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["font.family"] = "serif"
```

```python
dataset = pd.read_csv('UCS.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['P Wave'], color='orange')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('P Wave Velocity (m/s)', size=15)
plt.title('Comparative Analysis of Temperature and P Wave Velocity',position=(0.5, 0.945),size=11)
plt.savefig("UCS(C1)",dpi=1000,bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['Dt'], color='blue')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('Thermal Damage', size=15)
plt.title('Comparative Analysis of Temperature and Thermal Damage',position=(0.5, 0.945),size=11)
plt.savefig("UCS(C2)",dpi=1000,bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['Max stress σmax (MPa)'], color='green')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('Max stress σmax (MPa)', size=15)
plt.title('Comparative Analysis of Temperature and Maximum Stress',position=(0.5, 0.955),size=11)
plt.savefig("UCS(C3)",dpi=1000,bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['Avg (σm1, σm2)'], color='red')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('Memory Stress (MPa)', size=15)
plt.title('Comparative Analysis of Temperature and Memory Stress',position=(0.5, 0.955),size=11)
plt.savefig("UCS(C4)",dpi=1000,bbox_inches = 'tight')
plt.show()
```
