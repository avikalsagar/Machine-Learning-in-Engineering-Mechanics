```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["font.family"] = "serif"
```

```python
dataset = pd.read_csv('BTS.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['Max stress σmax (MPa)'], color='green')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('Max stress σmax (MPa)', size=15)
plt.title('Comparative Analysis of Temperature and Maximum Stress',position=(0.5, 0.955),size=11)
plt.savefig("BTS(C1)",dpi=1000,bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset['Temp'], dataset['Avg (σm1, σm2)'], color='red')
plt.xlabel('Temperature (°C)', size=15)
plt.ylabel('Memory Stress (MPa)', size=15)
plt.title('Comparative Analysis of Temperature and Memory Stress',position=(0.5, 0.955),size=11)
plt.savefig("BTS(C2)",dpi=1000,bbox_inches = 'tight')
plt.show()
```
